```python
!pip install urllib
!pip install bs4
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
my_url="https://en.wikipedia.org/wiki/Mobile_operating_system"
uClient=uReq(my_url)
page_html=uClient.read()
uClient.close()
```

```python

!pip install html_parser

page_soup=soup(page_html,"html.parser")
containers=page_soup.findAll("table",{"class":"wikitable"})
container=containers[1]
```

```python
print(containers[1])
```

```python

android=containers[1].findAll("a",{"title":"Android (operating system)"})

ios=containers[1].findAll("a",{"title":"IOS"})
kaios=containers[1].findAll("a",{"title":"KaiOS"})
windows=containers[1].findAll("a",{"title":"Windows 10 Mobile"})
blackberry=containers[1].findAll("a",{"title":"BlackBerry 10"})
symbian=containers[1].findAll("a",{"title":"Symbian"})
#tot_sm=containers[1].findAll("th",).text

print(android[0].text)

print(ios[0].text)
print(kaios[0].text)
print(windows[0].text)
print(blackberry[0].text)
print(symbian[0].text)

```
